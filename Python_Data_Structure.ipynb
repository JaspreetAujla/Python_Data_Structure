#Theory Questions

1- What are data structures, and why are they important?
- Data Structures is way of organizing and storing data in an efficient and organised manner. They help in managing and organizing data which reduces the complexity and increases the efficiency of a python code.

2- Explain the difference between mutable and immutable data types with examples?
- Mutable data types:- these data types can be changed after they are created. For example: Lists and Dictionary.
- immutable data types:- these data types cannot be changed after they are created. Example: String and Tuple

3- What are the main differences between lists and tuples in Python?
- Lists are initialised with the help of '[]' while Tuples are initialised with the help of '()'.
- Lists are mutable while Tuples are immutable.
- Lists are slow as compared to Tuples.

4- Describe how dictionaries store data?
- Dictionaries store data in the form of key-value pair where each unique key is associated with a value.
- Key can be a string, number or tuple while values can be of any data type.
- Systax:
  - dict = {"name":"Raj, "number":974268, "(1,2,3,4):"tuple"}

5- Why might you use a set instead of a list in Python?
- The main reason to use a Set instead of a List is that Set prevents the storage of duplicate values. So, whenever we need a set of unique value, we use a Set.

6- What is a string in Python, and how is it different from a list?
- String in Python is sequence of characters enclosed in a quotes("" or ''). It is mainly used to store text. String is immutable.
- where as List is a ordered collection of mutable objects enclosed in [].

7- How do tuples ensure data integrity in Python?
- Tuples ensures data integrity by being immutable. Once a tuple is created, it's elements cannot be modified, added or removed.

8- What is a hash table, and how does it relate to dictionaries in Python?
- A **hash table** is a structure that stores data as key–value pairs and uses a hash function to find values quickly.
- Python **dictionaries** are built using hash tables, which is why they give fast access to data.
When you use a dictionary key, Python converts it into a hash and finds the value quickly.
Only **hashable** (mostly immutable) objects can be used as dictionary keys.

9- Can lists contain different data types in Python?
- Yes, list is a heterogenous collection of elements so it can contain values of different data types.

10- Explain why strings are immutable in Python?
- Strings are immutable by design to keep them safe, consistent, and efficient. Immutability makes strings hashable (usable as dictionary keys), memory-efficient, and thread-safe, ensuring they can be reused without unexpected changes.

11- What advantages do dictionaries offer over lists for certain tasks?
- Dictionaries offer **faster lookups** because they access values using keys instead of searching through indexes. They allow you to use **meaningful keys**, making code easier to read and manage. They’re also better for storing **key–value relationships** and automatically prevent duplicate keys.

12- Describe a scenario where using a tuple would be preferable over a list.
- A tuple is preferable when you want to store data that should not change, such as coordinates (x, y), dates, or configuration values. Because tuples are immutable, they protect the data from accidental modification.
- A good example is storing student records like (name, age, roll_no) where the structure and values should stay fixed. Since the data isn’t meant to change, a tuple ensures it remains protected. Tuples are also good for returning multiple values from a function in a safe, immutable form.

13- How do sets handle duplicate values in Python?
- In Set, the duplicate element is automatically igonred. Hence, ensuring that all the elements in the Set are unique because Set dosen't allow duplicate values.

14- How does the “in” keyword work differently for lists and dictionaries?
- The **"in"** keyword works differently for lists and dictionaries:

  - In a **list**, "in" checks whether a **value** is present in the list.
  - In a **dictionary**, "in" checks only for the **keys**, not the values.
  So, 5 in [1,5,7] checks values, but 5 in {5: "A", 6: "B"} checks keys.

15- Can you modify the elements of a tuple? Explain why or why not.
- No, we can't modify the elements of a tuple because it is an immutable data structure which means once a Tuple is created, it's elements can't be modified, added or removed.

16- What is a nested dictionary, and give an example of its use case?
- A **nested dictionary** is a dictionary that contains another dictionary as its value, allowing you to store structured or grouped data.
- It's useful for representing complex information, like storing student details under each student ID.
Example: students = {101: {"name": "Raj", "age": 20}, 102: {"name": "Asha", "age": 19}}.

17- Describe the time complexity of accessing elements in a dictionary.
- Accessing elements in a dictionary has an average time complexity of O(1) because dictionaries use a hash table internally.
- This means Python can find the value directly using the key's hash.

18- In what situations are lists preferred over dictionaries?
- Lists are preferred over dictionaries when:
  - You need **ordered** data that you will access by position (index).
  - The data doesn't require key-value pairs—just a simple sequence.
  - You need fast operations like slicing, sorting, or iterating in order.
  - The items don't need unique identifiers, unlike dictionary keys.

19-  Why are dictionaries considered unordered, and how does that affect data retrieval?
- Dictionaries are considered unordered because they don’t store items based on a fixed position like lists do; instead, they store them based on hash values of the keys.
This means you can’t rely on items appearing in a specific order when retrieving or iterating over them.
- Because dictionaries are unordered, you cannot retrieve items by position like you do with lists. However, this does not affect key-based retrieval. Accessing a value using its key is still very fast and reliable because Python uses hashing to find it directly.

20- Explain the difference between a list and a dictionary in terms of data retrieval.
- In a **list**, data is retrieved using **index positions** (like list[0]), so you must know the item's location.
- In a **dictionary**, data is retrieved using **keys** (like dict["name"]), which makes access faster and more meaningful.
  This means lists rely on order, while dictionaries rely on unique keys for quick lookups.
